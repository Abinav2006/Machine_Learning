```python
import pandas as pd
df = pd.read_csv('carprices.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.scatter(df['Mileage'],df['Sell Price($)'], color = 'red', marker = '+')
```

```python
plt.scatter(df['Age(yrs)'],df['Sell Price($)'], color='red',marker='+')
plt.xlabel('Age')
plt.ylabel('Sell Price($)')
```

```python
x = df[['Mileage','Age(yrs)']]
y = df['Sell Price($)']
```

```python
x
```

```python
y
```

```python
 from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=0.2) #random_state=>same data for taining
```

```python
len(x_train)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict([[46000,4]])
```

```python
model.score(x_test,y_test)
```
